# Missing Values Imputation

There are multiple reasons behind why values are missing in a dataset.

<br> <br>

**1. Missing Completely at Random (MCAR)**: This means that the missingness is completely random and has no relation to any other observation.

Examples:

1. **Weather Effect on Survey Completion:**
Suppose you conduct a survey on people's favorite ice cream flavors. If some responses are missing purely because of unpredictable weather changes (like a sudden rainstorm) that have nothing to do with people's ice cream preferences or any other factors, then the missingness of the survey results can be considered MCAR.

2. **Technical Glitches:**
Imagine you are collecting data electronically, and due to a random technical glitch, some data is not recorded. If this glitch is truly random and is not related to any particular characteristic of the observations, then this missing data is MCAR.

<br><br>

**2. Missing at Random (MAR):**
Data is missing due to another observed variable. For instance, we may observe that the percentage of missing values differs significantly based on other variables.

Examples:

1. **Income and Age:**
Imagine you're conducting a survey about personal finances. You notice that respondents with higher incomes are less likely to disclose their exact age. However, within the group of high-income respondents, the probability of missing age data does not depend on the age itself. In this case, age data is MAR with respect to income. The missingness of age is related to a variable we've observed (income) but not the age value that's missing.

<br> <br>
**3. Missing NOT at Random (MNAR):** MNAR occurs when there is a definite pattern in the missing variable. However, it is unrelated to any feature we can observe in our data. In fact, this may depend on an unobserved feature.

Examples:

1. **Academic Surveys - Poor Grades:**
Students might be asked to report their grades for various subjects. Students who performed poorly in a particular subject might be less likely to disclose their grades for that subject. Here, the missingness of the grade data is directly related to the grade value, indicating MNAR.

2. **Employee Feedback Surveys - Job Satisfaction:**
Employees in a company are surveyed about their job satisfaction. Those who are extremely dissatisfied might choose not to respond to the survey at all, fearing repercussions or believing it won't lead to change. If the most dissatisfied employees are the ones not providing feedback, then their missing responses are MNAR.

Identifying the reason for missingness can be extremely useful for further analysis, imputation, and modeling.
